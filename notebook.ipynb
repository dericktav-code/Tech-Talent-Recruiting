```python
import pandas as pd
import re

# Load the resume dataset from a CSV file into a DataFrame
resumes = pd.read_csv('resumes.csv')
resumes.sample(3)
```

```python
job_title = r'^\s*([A-Za-z0-9\s,&/\'-]+?)\s{2,}(?:Summary|Experience|Skills|Highlights|Objective|Professional Profile|Executive Profile)'
```

```python
tech_skills = r'\b(Python|SQL|R|Excel)\b'
```

```python
education = r'\b(PhD|Master|Bachelor)\b'
degree_rank = ['PhD', 'Master', 'Bachelor']
```

```python
job_titles = []
for resume in resumes['Resume_str']:
    title_match = re.search(job_title, resume)
    job_titles.append(title_match.group(1).strip() if title_match else None)
```

```python
skills_list = []
for resume in resumes['Resume_str']:
    skills_found = re.findall(tech_skills, resume)
    skills_list.append(list(set(skills_found)) if skills_found else None)
```

```python
educations = []
for resume in resumes['Resume_str']:
    degrees_found = re.findall(education, resume)
    if degrees_found:
        educations.append(min(degrees_found, key=lambda d: degree_rank.index(d)))
    else:
        educations.append(None)
```

```python
candidates_df = pd.DataFrame({
    'id': resumes['ID'].astype('Int64'),
    'job_title': job_titles,
    'tech_skills': skills_list,
    'education': educations
})
```

```python
candidates_df = candidates_df.dropna(subset=['job_title', 'tech_skills', 'education'])
candidates_df = candidates_df[candidates_df['job_title'] != ""]
candidates_df = candidates_df.reset_index(drop=True)
candidates_df
```
